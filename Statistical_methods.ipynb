```python
import seaborn as sns 
import pandas as pd

df = pd.read_csv('US_Accidents_March23.csv')
df = df[['Severity', 'Start_Time', 'End_Time', 'Start_Lat', 'Start_Lng', 'End_Lat', 'End_Lng', 'Distance(mi)', 'Street', 'City', 'County', 'State', 'Zipcode', 'Temperature(F)','Wind_Chill(F)', 'Humidity(%)','Pressure(in)', 'Visibility(mi)', 'Wind_Speed(mph)', 'Precipitation(in)', 'Weather_Condition', 'Crossing', 'Junction', 'Stop', 'Traffic_Signal', 'Sunrise_Sunset']]
```

```python
import matplotlib.pyplot as plt
import numpy as np

average_severity_by_state = df.groupby('State')['Severity'].mean().reset_index()
average_temperature_by_state = df.groupby('State')['Temperature(F)'].mean().reset_index()

x = average_severity_by_state['Severity']
y = average_temperature_by_state['Temperature(F)']
plt.scatter(x, y)
z = np.polyfit(x, y, 1)
p = np.poly1d(z)

#add trendline to plot
plt.plot(x, p(x), color = 'green')
plt.show()
```
